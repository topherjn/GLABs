```python
340.4 Lists
2024.01.25 vlb

#Example of Lists.sort  
# A function that returns the length of the value:
def myFunc(e):
  return len(e)
cars = ['Ford', 'Mitsubishi', 'BMW', 'VW']
cars.sort(reverse=True, key=myFunc)
print(cars)
print(type(cars))

# Study Guide: Python Lists

# Key Terms:
# - List: A collection of elements in a specific order, mutable and denoted by square brackets [].
# - Element: An individual item or value stored in a list.
# - Index: The position of an element in the list. Indexing starts from 0.
# - Mutable: Lists can be modified after creation.
# - Iterable: Lists can be iterated over using loops or other methods.

# Creating Lists
# Lists can store elements of different data types, and elements can be duplicated.
# Syntax: list_name = [element1, element2, ...]

# Example:
fruits = ['apple', 'orange', 'banana', 'grapes']
# Here, we have a list called 'fruits' with four elements.

# Accessing Elements
# Elements in a list can be accessed using their index.
# Syntax: list_name[index]

# Example:
print(fruits[0])  # Output: 'apple'
print(fruits[2])  # Output: 'banana'

# Modifying Elements
# You can modify the elements in a list by assigning new values to their indices.
# Syntax: list_name[index] = new_value

# Example:
fruits[1] = 'kiwi'
print(fruits)  # Output: ['apple', 'kiwi', 'banana', 'grapes']

# List Slicing
# Slicing allows you to extract a portion of the list.
# Syntax: list_name[start_index:end_index]
# Note: The start_index element is included, but the end_index element is not.

# Example:
print(fruits[1:3])  # Output: ['kiwi', 'banana']

# Adding Elements
# You can add elements to the end of a list using the 'append()' method.
# Syntax: list_name.append(new_element)

# Example:
fruits.append('mango')
print(fruits)  # Output: ['apple', 'kiwi', 'banana', 'grapes', 'mango']

# Inserting Elements
# You can insert elements at a specific index using the 'insert()' method.
# Syntax: list_name.insert(index, new_element)

# Example:
fruits.insert(1, 'pear')
print(fruits)  # Output: ['apple', 'pear', 'kiwi', 'banana', 'grapes', 'mango']

# Removing Elements
# You can remove elements by their value using the 'remove()' method.
# Syntax: list_name.remove(element)

# Example:
fruits.remove('banana')
print(fruits)  # Output: ['apple', 'pear', 'kiwi', 'grapes', 'mango']

# You can also remove elements by their index using the 'pop()' method.
# Syntax: list_name.pop(index)

# Example:
fruits.pop(2)
print(fruits)  # Output: ['apple', 'pear', 'grapes', 'mango']

# List Length
# You can find the number of elements in a list using the 'len()' function.
# Syntax: len(list_name)

# Example:
print(len(fruits))  # Output: 4

# Checking for Element Existence
# You can check if an element exists in a list using the 'in' keyword.
# Syntax: element in list_name

# Example:
print('apple' in fruits)  # Output: True
print('kiwi' in fruits)   # Output: False

# Looping through a List
# You can use a 'for' loop to iterate over the elements in a list.
# Syntax: for element in list_name:

# Example:
for fruit in fruits:
    print(fruit)
# Output:
# 'apple'
# 'pear'
# 'grapes'
# 'mango'

# List Concatenation
# You can concatenate two lists using the '+' operator.
# Syntax: new_list = list1 + list2

# Example:
more_fruits = ['pineapple', 'watermelon']
all_fruits = fruits + more_fruits
print(all_fruits)  # Output: ['apple', 'pear', 'grapes', 'mango', 'pineapple', 'watermelon']

# List Sorting
# You can sort the elements in a list using the 'sort()' method.
# Syntax: list_name.sort()

# Example:
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]
numbers.sort()
print(numbers)  # Output: [1, 1, 2, 3, 4, 5, 5, 6, 9]

# List Reversing
# You can reverse the order of elements in a list using the 'reverse()' method.
# Syntax: list_name.reverse()

# Example:
numbers.reverse()
print(numbers)  # Output: [9, 6, 5, 5, 4, 3, 2, 1, 1]

# List Comprehension
# List comprehension provides a concise way to create lists based on existing lists.
# Syntax: new_list = [expression for item in list_name if condition]

# Example:
squared_numbers = [num ** 2 for num in numbers if num % 2 == 0]
print(squared_numbers)  # Output: [36, 4]

# Conclusion:
# Lists are a fundamental data structure in Python, allowing you to store and manipulate collections of elements.
# Understanding lists and their various methods will significantly enhance your Python programming skills.


```
